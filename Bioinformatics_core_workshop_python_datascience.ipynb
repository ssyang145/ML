#### Ploting and Basic Statistics

```python
# Create a boxplot
df.boxplot('mean radius',by='target')
```

```python
# Create two new dataframes
malignant = df['mean radius'][df['target'] == 0]
benign = df['mean radius'][df['target'] == 1]
```

```python
malignant
```

```python
# T-test
from scipy.stats import ttest_ind, ttest_ind_from_stats
t,p=ttest_ind(malignant,benign, equal_var=False)
print ('t-test p-value for mean radius of breat cancer ',p)
```

```python
df.plot.scatter('mean radius','mean symmetry',c=df['target'],colormap='viridis')
```

#### Build logistic regression with the two features to predict cancer type

```python
# Build X and y matrix
X=df[['mean radius','mean symmetry']]
y=df['target']
```

```python
# Seperate data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0) # test_size=0.2
```

```python
# train logistic regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression().fit(X_train, y_train)
lr.score(X_test,y_test)
```

```python
# Use all features
X=df.drop(['target'],axis=1) # drop column 
y=df['target']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0) # test_size=0.2
lr = LogisticRegression().fit(X_train, y_train)
lr.score(X_test,y_test)
```

#### Principal Components Analysis (PCA)

```python
# Before applying PCA, each feature should be centered (zero mean) and with unit variance
from sklearn.preprocessing import StandardScaler
X_normalized = StandardScaler().fit(X).transform(X) 
```

```python
# Use PCA to find the first two principal components
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2).fit(X_normalized)
X_pca = pca.transform(X_normalized)
```

```python
print (X.shape,X_normalized.shape,X_pca.shape) # check dimension
```

```python
# plot the PCA-transformed version of the breast cancer dataset
plt.figure(figsize=(10,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=df['target'], cmap='plasma')
plt.xlabel = "First Principal Component"
plt.ylabel = "Second Principal component"
plt.show()
```

```python
# check contribution of each feature to PCs
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
%matplotlib inline
plt.figure(figsize=(10,6))
sns.heatmap(df_comp, cmap='plasma')
```
