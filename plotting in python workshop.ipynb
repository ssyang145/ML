### Matplotlib Tutorial Overview

```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Prepare the data
x = np.linspace(0, 10, 100)

# Plot the data
plt.plot(x, x, label='linear')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

What you can’t see on the surface is that you have -maybe unconsciously- made use of the built-in defaults that take care of the creation of the underlying components, such as the Figure and the Axes. 


In essence, there are two big components that you need to take into account:

    The Figure is the overall window or page that everything is drawn on. It’s the top-level component of all the ones that you will consider in the following points. You can create multiple independent Figures. A Figure can have several other things in it, such as a suptitle, which is a centered title to the figure. You’ll also find that you can add a legend and color bar, for example, to your Figure.
    
    To the figure you add Axes. The Axes is the area on which the data is plotted with functions such as plot() and scatter() and that can have ticks, labels, etc. associated with it. This explains why Figures can contain multiple Axes.


```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
plt.xlim(0.5, 4.5)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
ax.set_xlim(0.5, 4.5)
plt.show()
```

Tip: However, if you have multiple axes, it’s still better to make use of the first code chunk because it’s always better to prefer explicit above implicit code! 

Each Axes has an x-axis and a y-axis, which contain ticks, which have major and minor ticklines and ticklabels. There’s also the axis labels, title, and legend to consider when you want to customize your axes, but also taking into account the axis scales and gridlines might come in handy.


###  Creating your plot

```python
# Create a Figure
fig = plt.figure()

# Set up Axes
ax = fig.add_subplot(111)

# Scatter the data
ax.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))

# Show the plot
plt.show()
```

111 is equal to 1,1,1, which means that you actually give three arguments to add_subplot(). The three arguments designate the number of rows (1), the number of columns (1) and the plot number (1). So you actually make one subplot.

In most cases, you’ll use add_subplot() to create axes; Only in cases where the positioning matters, you’ll resort to add_axes(). Alternatively, you can also use subplots() if you want to get one or more subplots at the same time. You’ll see an example of how this works in the next section.

```python
from scipy import integrate
```

```python
A=1
f=1
```

```python
w=2*np.pi*f
```

```python
t = np.linspace(0, 1/f, 101)
a=A*np.sin(w*t)
```

```python
v = integrate.cumtrapz(a,t,initial=0)
```

```python
u= integrate.cumtrapz(v,t,initial=0)
```

```python
plt.figure(figsize=(9,6))
plt.plot(t,a, 'b*-', label="a (m/s2)")
plt.plot(t,v, 'ro-', label="v (m/s)")
plt.plot(t,u, 'g--', label="u (m)")
plt.xlabel('t (s)')
plt.ylabel('m')
plt.title('automation plots')
plt.legend(['a', 'v', 'u'])
plt.grid(True)
```

```python
A=1
u_list_f=[]
u_list_max=[]
plt.figure(figsize=(9.8,15), dpi=100)
for i,f in enumerate ([0.01,0.1,0.5,1,2,10]):
    w=2*np.pi*f
    t = np.linspace(0, 1/f, 101)
    a=A*np.sin(w*t)
    v = integrate.cumtrapz(a,t,initial=0)
    u= integrate.cumtrapz(v,t,initial=0)
    u_list_f.append(u[-1])
    u_list_max.append(max(u))
    print ("last u =",u[-1],"when f=",f)
    plt.subplot(3,2,i+1)
    plt.plot(t,a, 'b*-', label=u"a (m/s2)")
    plt.plot(t,v, 'ro-', label=u"v (m/s)")
    plt.plot(t,u, 'g--', label=u"u (m)")
    plt.xlabel('t (s)')
    plt.ylabel('m')
    plt.legend(['a', 'v', 'u'])
    plt.title("f="+ str(f))
    plt.grid(True)
```

```python
A=1
u_list_f=[]
u_list_max=[]
plt.figure(figsize=(9.8,15), dpi=100)
for i,f in enumerate ([0.01,0.1,0.5,1,2,10]):
    w=2*np.pi*f
    t = np.linspace(0, 1, 101)
    a=A*np.sin(w*t)
    v = integrate.cumtrapz(a,t,initial=0)
    u= integrate.cumtrapz(v,t,initial=0)
    u_list_f.append(u[-1])
    u_list_max.append(max(u))
    print ("last u =",u[-1],"when f=",f)
    plt.subplot(3,2,i+1)
    #plt.plot(t,a, 'b*-', label=u"a (m/s2)")
    #plt.plot(t,v, 'ro-', label=u"v (m/s)")
    plt.plot(t,u, 'g--', label=u"u (m)")
    plt.xlabel('t (s)')
    plt.ylabel('m')
    plt.legend(['a', 'v', 'u'])
    plt.title("f="+ str(f))
    plt.grid(True)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.plot([0.01,0.1,0.5,1,2,10], u_list_f,'g-*')
plt.xlabel('f')
plt.ylabel('max u')
plt.title('relationship between f and final u')
plt.subplot(1,2,2)
plt.plot([0.01,0.1,0.5,1,2,10], u_list_max,'r-o')
plt.xlabel('f')
plt.ylabel('final u')
plt.title('relationship between f and max u')
```

## Python Seaborn

Seaborn, which provides a high-level interface to draw statistical graphics.

Seaborn extends Matplotlib and that’s why it can address the two biggest frustrations of working with Matplotlib. Or, as Michael Waskom says in the “introduction to Seaborn”: “If matplotlib “tries to make easy things easy and hard things possible”, seaborn tries to make a well-defined set of hard things easy too.”

One of these hard things or frustrations had to do with the default Matplotlib parameters. 


Another frustration of users more apparent: the fact that working with DataFrames doesn’t go quite as smoothly with Matplotlib, which can be annoying if you’re doing exploratory analysis with Pandas. And that’s exactly what Seaborn addresses: the plotting functions operate on DataFrames and arrays that contain a whole dataset.






### Loading data to construct Seaborn plots

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris data
iris = sns.load_dataset("iris")

# Construct iris plot
sns.catplot(x="species", y="petal_length", data=iris,kind="violin")

# Show plot
plt.show() 
# Matplotlib still underlies Seaborn, which means that the anatomy of the plot is still the same and that you’ll need to use plt.show() to make the image appear to you.
```

```python
# Load data
titanic = sns.load_dataset("titanic")

# Set up a factorplot
g = sns.catplot("class", "survived", "sex", data=titanic, kind="bar", palette="muted", legend=False)
                   
# Show plot
plt.show()
```

### Scaling Seaborn plots

```python
# Set context to `"paper"`
sns.set_context("paper", font_scale=3, rc={"font.size":8,"axes.labelsize":5})

# Load iris data
iris = sns.load_dataset("iris")

# Construct iris plot
sns.swarmplot(x="species", y="petal_length", data=iris)

# Show plot
plt.show()
```

```python
# Initialize Figure and Axes object
fig, ax = plt.subplots(figsize=(10,4))

# Load in the data
iris = sns.load_dataset("iris")

# Create swarmplot
sns.swarmplot(x="species", y="petal_length", data=iris, ax=ax)

# Show plot
plt.show()
```

```python
# Load data
titanic = sns.load_dataset("titanic")

# Set up a factorplot
g = sns.catplot("class", "survived", "sex", data=titanic, kind="bar", height=6, aspect=2, palette="muted", legend=False)


# Show plot
plt.show()
```

### Setting log scale

```python
import pandas as pd

# Create the data
x = 10 ** np.arange(1, 10)
y = x * 2
data = pd.DataFrame(data={'x': x, 'y': y})

# Initialize figure and ax
fig, ax = plt.subplots()

# Set the scale of the x-and y-axes
ax.set(yscale="log")

# Set the `xlim`
#ax.set(xlim=(0, 100))

# Set title
ax.set_title("reg plot log scale")

# Create a regplot
sns.regplot("x", "y", data, ax=ax, scatter_kws={"s": 100})

# Show plot
plt.show()
```
